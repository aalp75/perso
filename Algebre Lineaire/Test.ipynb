```python
from LinearAlgebra import *
import timeit
```

```python
s = 10
A = np.random.randint(-4,4,size=(s,s))
print("size :",A.shape)
print(A)
S = np.dot(A.T,A)
```

```python
print("Test if A*A.T is symetric :")
print(is_sym(S))
```

```python
print("For matrix",A.shape)
start = timeit.default_timer()
detA = determinant_naive(A)
stop = timeit.default_timer()
print("determinant of A (naive method):",detA)
print("time :",stop-start,"secs")
start = timeit.default_timer()
detA = determinant_gauss(A)
stop = timeit.default_timer()
print("determinant of A (gaussian method):",detA)
print("time :",stop-start,"secs")
start = timeit.default_timer()
detA = np.linalg.det(A)
stop = timeit.default_timer()
print("determinant of A (numpy linalg mehtod):",detA)
print("time :",stop-start,"secs")
```

```python
print("Test if A*A.T is positive definite")
print(is_def_pos_sylvester(S))
```

```python
print("Test the inverse of A with gaussian with A*inv(A) == Id")
invA = inverse_gauss(A)
print(np.round(np.dot(invA,A)+0.0001,3))
```

```python
print("Test Cholesky algorithme on S=A*A.T")
L = cholesky(S)
if (np.linalg.norm(np.dot(L,L.T)-S)<0.001):
    print(True)
else :
    print(False)
```
